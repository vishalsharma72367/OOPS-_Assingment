1. What are the five key concepts of Object-Oriented Programming (OOP)?

Ans.The five key concepts of Object-Oriented Programming (OOP) are:


1. Encapsulation: This concept involves bundling data and methods that operate on that data within a single unit, called a class or object. This helps protect the data from external interference and misuse.


1. Abstraction: Abstraction involves exposing only necessary information to the outside world while hiding implementation details. This helps reduce complexity and improve modularity.


1. Inheritance: Inheritance allows one class to inherit properties and behavior from another class, facilitating code reuse and hierarchy.


1. Polymorphism: Polymorphism enables objects of different classes to be treated as objects of a common superclass, allowing for more flexibility in programming.


1. Composition: Composition involves combining objects or classes to form new objects or classes, promoting modularity and reuse.


These concepts form the foundation of OOP and enable developers to create robust, scalable, and maintainable software systems.


Additional Key Concepts:

- Class: A blueprint or template defining properties and behavior.
- Object: An instance of a class.
- Interface: A contract specifying methods that must be implemented.


Benefits of OOP:

- Modularity
- Reusability
- Easier maintenance
- Improved readability
- Scalability

Ans. In Python, instance methods and class methods are two types of methods that can be defined inside a class.


Instance Methods:

Instance methods are methods that belong to an instance of a class. They operate on the instance's attributes and are used to perform actions related to that specific instance.


Characteristics:


- Defined inside a class
- First parameter is self (a reference to the instance)
- Can access and modify instance attributes
- Called on an instance of the class


Example:

        self.name = name
        self.age = age

    def greet(self):

Class methods are methods that belong to a class itself, rather than an instance of the class. They operate on class attributes and are used to perform actions related to the class.


Characteristics:


- Defined inside a class with the @classmethod decorator
- First parameter is cls (a reference to the class)
- Can access and modify class attributes
- Called on the class itself


Example:

Characteristics:


- Defined inside a class with the @staticmethod decorator
- No implicit first parameter (self or cls)
- Cannot access or modify class or instance attributes

4. How does Python implement method overloading? Give an example.

Ans. Method Overloading:

Method overloading is a feature that allows multiple methods with the same name to be defined, but with different parameter lists.


Python's Approach:

Python achieves method overloading through:


1. Default argument values
2. Variable number of arguments (*args and **kwargs)
3. Single Dispatch (using @singledispatch decorator from functools module)


Example 1: Using Default Argument Values


class Calculator:
    def calculate(self, a, b, operation='add'):
        if operation == 'add':
            return a + b
        elif operation == 'subtract':
            return a - b
        else:
            raise ValueError("Invalid operation")


calculator = Calculator()
print(calculator.calculate(5, 3))  # Output: 8
print(calculator.calculate(5, 3, 'subtract'))  # Output: 2



Example 2: Using Variable Number of Arguments


class Logger:
    def log(self, *args):
        if len(args) == 1:
            print(f"Logging: {args[0]}")
        elif len(args) > 1:
            print(f"Logging: {args}")


logger = Logger()
logger.log("Hello")
logger.log("Hello", "World")



Example 3: Using Single Dispatch


from functools import singledispatch

@singledispatch
def fun(arg):
    return arg


@fun.register
def _(arg: int):
    return "Integer"


@fun.register
def _(arg: str):
    return "String"


print(fun(5))  # Output: Integer
print(fun("Hello"))  # Output: String





5. What are the three types of access modifiers in Python? How are they denoted?

Ans. Python has three types of access modifiers:


1. Public
2. Protected
3. Private


Denotation:

- Public: No specific notation (e.g., variable)
- Protected: Single underscore prefix (e.g., _variable)
- Private: Double underscore prefix (e.g., __variable)


Access Levels:

- Public: Accessible from anywhere
- Protected: Accessible within class and subclasses
- Private: Accessible only within class (using name mangling)


Example:

class MyClass:
    def __init__(self):
        self.public_var = 10  # Public
        self._protected_var = 20  # Protected
        self.__private_var = 30  # Private

    def print_private(self):
        print(self.__private_var)


obj = MyClass()

# Accessing public variable
print(obj.public_var)  # Output: 10

# Accessing protected variable (not recommended)
print(obj._protected_var)  # Output: 20

# Accessing private variable (not recommended)
# print(obj.__private_var)  # Raises AttributeError
print(obj._MyClass__private_var)  # Output: 30 (using name mangling)

# Accessing private variable through method
obj.print_private()  # Output: 30



Best Practices:

- Use public variables for data that needs to be accessed externally.
- Use protected variables for data that should be accessed within class and subclasses.
- Use private variables for sensitive or internal data.

7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

Ans. Method Resolution Order (MRO)


In Python, the Method Resolution Order (MRO) is the order in which the interpreter searches for methods and attributes in a class's inheritance hierarchy. It's used to resolve conflicts when multiple classes define the same method or attribute.


How MRO Works


1. The interpreter starts searching from the current class.
2. If not found, it searches in the parent classes (depth-first, left-to-right).
3. If still not found, it searches in the parent classes of the parent classes, and so on.


Retrieving MRO Programmatically


You can retrieve the MRO of a class using the mro() method:



class A:
    pass

class B(A):
    pass

class C(B):
    pass

print(C.mro())  # Output: [<class '__main__.C'>, <class '__main__.B'>, <class '__main__.A'>, <class 'object'>]



Alternatively, you can use the __mro__ attribute:



print(C.__mro__)  # Output: (<class '__main__.C'>, <class '__main__.B'>, <class '__main__.A'>, <class 'object'>)



Example with Multiple Inheritance



class A:
    def method(self):
        print("A's method")

class B:
    def method(self):
        print("B's method")

class C(A, B):
    pass

print(C.mro())  # Output: [<class '__main__.C'>, <class '__main__.A'>, <class '__main__.B'>, <class 'object'>]

c = C()
c.method()  # Output: A's method



In this example, the MRO determines that A's method is called instead of B's method.


C3 Linearization


Python's MRO uses the C3 linearization algorithm to ensure that:


1. Local precedence ordering is preserved.
2. Monotonicity is preserved.


This ensures that the MRO is predictable and consistent, even in complex inheritance scenarios.

class Vector:

        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

import time
from functools import wraps


def timer_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Function '{func.__name__}' executed in {execution_time:.4f} seconds.")
        return result
    return wrapper


# Example usage
@timer_decorator
def example_function():
    time.sleep(1)  # Simulate some work
    print("Function executed.")


example_function()


# Output:
# Function executed.
# Function 'example_function' executed in 1.0011 seconds.



This decorator:


1. Uses the time module to measure the execution time.
2. Wraps the original function using @wraps from functools to preserve its metadata.
3. Prints the execution time after the function finishes.


You can apply this decorator to any function to measure its execution time.


Variations:


- Measure execution time in milliseconds: execution_time * 1000.
- Measure execution time in minutes: execution_time / 60.
- Log execution time instead of printing: replace print with logging function.


Best practices:


- Use @wraps to preserve original function metadata.
- Use meaningful decorator names.
- Document decorator behavior.

13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

Ans. The Diamond Problem is a challenge in multiple inheritance where a class inherits conflicting methods or attributes from its parent classes.


Diamond Inheritance Graph:



  A
 / \
B   C
 \ /
  D



Problem:


Class D inherits from classes B and C, which both inherit from class A. If class A has a method or attribute that classes B and C override, class D receives conflicting methods or attributes.


Python's Solution:


Python resolves the Diamond Problem using:


1. Method Resolution Order (MRO): Determines the order classes are searched for methods and attributes.
2. C3 Linearization: Ensures MRO consistency.


MRO Rules:


1. Local precedence ordering: Class D's methods and attributes take precedence.
2. Parent classes are searched in the order listed in the class definition.
3. If a parent class is searched, its subclasses are skipped.


Example:



class A:
    def method(self):
        print("A's method")

class B(A):
    def method(self):
        print("B's method")

class C(A):
    def method(self):
        print("C's method")

class D(B, C):
    pass

d = D()
d.method()  # Output: B's method



How Python Resolves Conflicts:


1. Searches class D for method.
2. If not found, searches class B.
3. Finds method in class B and uses it.


Best Practices:


1. Avoid deep inheritance.
2. Use multiple inheritance judiciously.
3. Ensure clear method and attribute naming.


Python's MRO Algorithm:


To view the MRO for a class, use the mro() method:



print(D.mro())
# Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]

        super().__init__(name, bases, attrs)

        cls.instances += 1

    return cls

15. Implement a static method in a class that checks if a given year is a leap year.

Ans. Here's an implementation of a static method in a class that checks if a given year is a leap year:



class DateUtil:
    @staticmethod
    def is_leap_year(year):
        """
        Checks if a given year is a leap year.

        Args:
            year (int): The year to check.

        Returns:
            bool: True if the year is a leap year, False otherwise.
        """
        return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)


# Example usage
print(DateUtil.is_leap_year(2020))  # Output: True
print(DateUtil.is_leap_year(2019))  # Output: False



Explanation:


- The is_leap_year method is decorated with @staticmethod to indicate it's a static method.
- The method takes an integer year as input.
- It returns True if the year is a leap year and False otherwise.


Leap Year Rules:


- A year is a leap year if it's divisible by 4.
- However, years divisible by 100 are not leap years unless they're also divisible by 400.


Benefits:


- Static methods can be called without creating an instance of the class.
- Useful for utility functions that don't depend on instance state.


Alternative Implementation:



class DateUtil:
    @classmethod
    def is_leap_year(cls, year):
        # Same implementation as above
        return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)



Note:


- @classmethod can also be used instead of @staticmethod.
- However, @staticmethod is more suitable for utility functions that don't rely on class or instance state.
